## **THEORY QUESTIONS**


### Question 1: What is unsupervised learning in the context of machine learning?

Unsupervised learning is a type of machine learning where the algorithm is trained on data without labeled outputs. It is used to identify hidden patterns or structures in the data, such as clustering data points into groups without predefined categories.


### Question 2: How does K-Means clustering algorithm work?

K-Means clustering works by initializing a specified number of cluster centers (k) randomly. Each data point is assigned to the nearest cluster center and then cluster centers are updated as the mean of all assigned points. This process repeats until the cluster centers no longer change significantly.

### Question 3: Explain the concept of a dendrogram in hierarchical clustering.

A dendrogram is a tree-like diagram that shows the hierarchical arrangement of clusters produced by hierarchical clustering. It illustrates how individual points or clusters are progressively merged based on similarity until a single cluster remains.


### Question 4: What is the main difference between K-Means and Hierarchical Clustering?

The main difference is that K-Means requires specifying the number of clusters in advance, while Hierarchical Clustering does not. Hierarchical produces a dendrogram allowing flexibility to choose the number of clusters later.


### Question 5: What are the advantages of DBSCAN over K-Means?

DBSCAN does not require specifying the number of clusters and can identify arbitrarily shaped clusters. It is also able to detect noise (outliers), unlike K-Means, which always assigns every point to a cluster.

DBSCAN does not require specifying the number of clusters and can identify arbitrarily shaped clusters. It is also able to detect noise (outliers), unlike K-Means, which always assigns every point to a cluster.


### Question 6: When would you use Silhouette Score in clustering?

Silhouette Score is used to evaluate the quality of clustering by measuring how similar each data point is to its own cluster compared to other clusters. Higher scores indicate better-defined clusters and help decide the optimal number of clusters.


### Question 7: What are the limitations of Hierarchical Clustering?

Hierarchical clustering doesnâ€™t scale well for large datasets because it requires computing and storing a large distance matrix. Also, once clusters are merged, they cannot be split again, limiting flexibility.


### Question 8: Why is feature scaling important in clustering algorithms like K-Means?

Feature scaling is important because K-Means calculates distances between points. If features have different scales, some will dominate the distance calculations, leading to biased clustering. Scaling ensures fair contribution from all features.


### Question 9: How does DBSCAN identify noise points?

DBSCAN labels points as noise if they do not belong to any dense region. Specifically, points with fewer than a minimum number of neighbors within a given radius (epsilon) are marked as noise.


### Question 10: Define inertia in the context of K-Means.

Inertia is the sum of squared distances of each data point to its assigned cluster center. Lower inertia indicates that data points are closer to their centroids, implying tighter clusters.


### Question 11: What is the elbow method in K-Means clustering?

The elbow method helps determine the optimal number of clusters by plotting inertia versus the number of clusters. The 'elbow' point where inertia reduction slows suggests the best k value.


### Question 12: Describe the concept of "density" in DBSCAN.

Density in DBSCAN refers to the number of points within a given radius. A point belongs to a cluster if it lies in a dense area where the number of neighboring points exceeds a threshold.


### Question 13: Can hierarchical clustering be used on categorical data?

Yes, hierarchical clustering can be applied to categorical data using appropriate distance measures such as Hamming distance instead of Euclidean distance.


### Question 14: What does a negative Silhouette Score indicate?

A negative Silhouette Score indicates that a data point is likely assigned to the wrong cluster, being closer to another cluster than its own.


### Question 15: Explain the term "linkage criteria" in hierarchical clustering.

Linkage criteria determine how distances between clusters are calculated. Common methods include single linkage (minimum distance), complete linkage (maximum distance), and average linkage (average distance).


### Question 16: Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

K-Means assumes spherical clusters of similar size. It performs poorly when clusters vary significantly in size or density because it tends to split large clusters or merge small ones incorrectly.


### Question 17: What are the core parameters in DBSCAN, and how do they influence clustering?

The core parameters are 'epsilon' (eps) and 'min_samples'. Eps defines the neighborhood radius, and min_samples is the minimum points required to form a cluster. Too small eps increases noise, while too large eps may merge clusters incorrectly.


### Question 18: How does K-Means++ improve upon standard K-Means initialization?

K-Means++ spreads initial cluster centers more strategically to avoid poor initialization. This helps the algorithm converge faster and leads to better clustering results.


### Question 19: What is agglomerative clustering?

Agglomerative clustering is a hierarchical clustering method where each data point starts as an individual cluster, and clusters are merged iteratively based on similarity until a single cluster remains.


### Question 20: What makes Silhouette Score a better metric than just inertia for model evaluation?

Silhouette Score accounts for both cohesion (intra-cluster similarity) and separation (inter-cluster dissimilarity), providing a more balanced evaluation of clustering quality, whereas inertia only measures compactness.

### Question 21: Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X)

# Plotting
plt.figure(figsize=(8,6))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=clusters, palette='viridis')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='red', marker='X')
plt.title('K-Means Clustering with 4 Centers')
plt.show()
```

### Question 29: Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
sil_scores = []
k_values = [2, 3, 4, 5]

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    sil_scores.append(score)

# Plotting silhouette scores
plt.figure(figsize=(8,6))
sns.barplot(x=k_values, y=sil_scores, palette='viridis')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k')
plt.show()
```

### Question 30: Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

### Question 31: Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic overlapping data
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)

# Plot decision boundaries
h = 0.1  # mesh step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.4, cmap='viridis')
sns.scatterplot(x=X[:,0], y=X[:,1], hue=labels, palette='tab10')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='red', marker='X')
plt.title('K-Means Clustering with Decision Boundaries')
plt.show()
```

### Question 34: Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Compute inertia for K = 2 to 6
k_values = range(2, 7)
inertia = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plotting
plt.figure(figsize=(8,6))
sns.lineplot(x=k_values, y=inertia, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Inertia vs K for Breast Cancer Data')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
wine = load_wine()
X = wine.data
feature_names = wine.feature_names[:4]

# Cluster
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# DataFrame for seaborn
df = pd.DataFrame(X[:, :4], columns=feature_names)
df["Cluster"] = labels

# Pairplot
sns.pairplot(df, hue="Cluster", palette="tab10")
plt.show()
```

### Question 47: Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate noisy blobs
X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

# DBSCAN clustering
db = DBSCAN(eps=1.2, min_samples=10).fit(X)
labels = db.labels_

# Count clusters & noise
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print("Clusters found:", n_clusters)
print("Noise points:", n_noise)
```

### Question 48: Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

# Load Digits dataset
digits = load_digits()

# t-SNE
X_tsne = TSNE(n_components=2, random_state=42).fit_transform(digits.data)

# Agglomerative clustering
agg = AgglomerativeClustering(n_clusters=10)
labels = agg.fit_predict(X_tsne)

# Plot
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=labels, cmap='tab10')
plt.title("Digits Dataset - t-SNE + Agglomerative Clustering")
plt.show()
```
